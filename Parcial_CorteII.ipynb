```python
# Creamos un gráfico de dispersión (scatter plot) entre dos variables, azúcar residual & acidez volátil donde podemos ver que existen una gran dispersion, sinembargo los datos se concentran en una tasa de azucar residual de rango 0 a 20 en su mayoria y unos pocos salen de ella. y la acidez se dispersa sobre este rango mayoritario entre 0.1 a 0.7 princiaplmente
sns.scatterplot(x='residual sugar', y='volatile acidity', data=df)
plt.show()
```

```python
# Creamos un diagrama de caja utilizando la función boxplot() de Seaborn y especificando las variables que queremos analizar. Luego, utilizamos la función show() de Matplotlib para mostrar el gráfico.
sns.boxplot(x='quality', y='pH', data=df)
plt.show()
```

```python
# El diagrama de caja nos permitiría analizar la distribución de la densidad para cada calidad de vino. Podríamos utilizar este gráfico para identificar si existe una relación entre la calidad del vino y su densidad, como por ejemplo si la calidad del vino aumenta o disminuye a medida que la densidad se hace más alta o más baja.
sns.boxplot(x="quality", y="density", data=df)
```
